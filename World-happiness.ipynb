# Soal 4: Visualisasi lineplot Life Ladder negara Indonesia

*   Gunakan library seaborn untuk plotting line plot kolom Life Ladder dari hasil penggabungan negara Indonesia tahun 2006-2021 (soal 3) seperti expected output, jika tidak bisa menyelesaikan soal 3 silahkan download data berikut 
[Download.csv](https://drive.google.com/uc?export=download&id=19g5oURnRNlbXPUEQQyvimoEx8DwqMPLW) 
*   Berikan insight dari hasil visualisasi tersebut

```python
#code here
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df1 = pd.read_csv('world-happiness-report.csv')
df2 = pd.read_csv('world-happiness-report-2021.csv')
sns.set_theme(style="whitegrid")
df1_indonesia = df1[df1['Country name']== 'Indonesia']
df1_indo1 = df1_indonesia.reset_index(drop = True)
df1_indo2 = df1_indo1.drop(['Positive affect', 'Negative affect'], axis = 1)

dfadd = df2.insert(1, 'year', 2021, True)
df2_indonesia = df2[df2['Country name'] == 'Indonesia']
df2_1 = df2_indonesia.drop(['Regional indicator', 'Standard error of ladder score', 'upperwhisker', 'lowerwhisker', 'Ladder score in Dystopia', 'Explained by: Log GDP per capita', 'Explained by: Social support', 'Explained by: Healthy life expectancy', 'Explained by: Freedom to make life choices', 'Explained by: Generosity', 'Explained by: Perceptions of corruption', 'Dystopia + residual'], axis = 1)
df2_1_rename = df2_1.rename(columns = {'Ladder score':'Life Ladder',
 'Logged GDP per capita':'Log GDP per capita',
  'Healthy life expectancy':'Healthy life expectancy at birth'})

df_1_2 = [df1_indo2, df2_1_rename]
gabungan = pd.concat(df_1_2)
hasil = gabungan.reset_index(drop=True)

grafik = hasil[['year', 'Life Ladder']]
sns.lineplot(data = grafik, x = 'year', y = 'Life Ladder', marker = 'o')

plt.title('Indonesia\'s ladder score over the years')
```

# Soal 6: Visualisasi Scatter plot

Buatlah visualisasi scatter plot (perpaduan seaborn dan matplotlib) dari df2 antara GDP per capita ('Logged GDP per capita') dan  skor korupsi ('Perceptions of corruptions')  sesuai expected output dengan ketentuan berikut:

Gunakan kolom skor kebebasan beraksi ('Freedom to make life choices') sebagai ukuran pointnya kalikan suatu angka yang cukup besar agar terlihat.

Gunakan kolom skor kesehatan ('Healthy life expectancy') sebagai warna point, berikan cmap 'plasma'.

Gunakan regplot pada seaborn (sns.regplot) dengan argumen scatter=False dan color = 'black'

```python
#code here
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df2 = pd.read_csv('world-happiness-report-2021.csv')
plt.figure(figsize = (4,2))

plt.scatter(data=df2, x=df2['Logged GDP per capita'], y=df2['Perceptions of corruptions'], s=df2['Freedom to make life choices'], c=df2['Healthy life expectancy'], alpha=0.6)
plt.colorbar()
plt.title('Corruption vs GDP percapita')
plt.xlabel('Logged GDP per capita')
plt.ylabel('Perceptions of corruptions')
sns.regplot(data = df2, x = 'Logged GDP per capita', y= 'Perceptions of corruptions')

plt.show()
```
